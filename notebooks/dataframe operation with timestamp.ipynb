```python
from datetime import datetime, date, timedelta, timezone
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('../data/020_intermediate/messages.csv')
df
```

```python
# timezon jst
JST = timezone(timedelta(hours=+9), 'JST')
# now at JST
dt_unix_org = datetime.fromtimestamp(0, JST)
dt_now_jst = datetime.now(JST)
dt_now_jst
```

```python
term_lim = timedelta(days=8)
for idx, row in df.iterrows():
    dt_rec_jst = datetime.fromtimestamp(row.timestamp, JST)
    diff_dt_from_now = dt_now_jst - dt_rec_jst
    print('now: ',dt_now_jst)
    print('record ts: ', dt_rec_jst)
    print('now - rects: ', diff_dt_from_now, type(diff_dt_from_now))
    print('now - rects < 8days? = ', (diff_dt_from_now < term_lim))
    print('===========================')
```

```python
td_8days_sec = timedelta(days=8)
td_8days_sec.total_seconds()
```

```python
now_in_sec = (dt_now_jst - dt_unix_org).total_seconds()
days8_in_sec = td_8days_sec.total_seconds()
df_lastweek = df.query('@now_in_sec - timestamp < @days8_in_sec')
```
