```python
import sys
sys.path.append('../src')
from d000_utils import slack_app as sa
import json
```

```python
# load api token
credentials_root = '../conf/local/'
credential_fpath = credentials_root + 'credentials.json'
print('load credential.json ...')
with open(credential_fpath, 'r') as f:
    credentials = json.load(f)
```

```python
# start slack app
print('start slack app ...')
app = sa.SlackApp(
    credentials['channel_api_key'],
    credentials['user_api_key']
    )
```

```python
# list channel info
app.list_channel_names()
```

```python
app.get_usr_info()
```

```python
# ========================================
# Load Raw User Info
# ========================================
import requests
payload = {'token': credentials['user_api_key']}
response = requests.get('https://slack.com/api/users.list', params=payload)
if response.status_code == 200:
    json_data = response.json()
    if 'members' in json_data.keys():
        members = json_data['members']
    with open('user_info.json', 'w') as f:
        json.dump(members, f, indent=4, ensure_ascii=False)
```

```python
# ========================================
# Load Raw Channels Info
# ========================================
payload = {'token': credentials['channel_api_key']}
response = requests.get('https://slack.com/api/channels.list', params=payload)
if response.status_code == 200:
    json_data = response.json()
    if 'channels' in json_data.keys():
        channels = json_data['channels']
    with open('channel_info.json', 'w') as f:
        json.dump(channels, f, indent=4, ensure_ascii=False)
```

```python
channel_id_list = []
for ch in channels:
    channel_id_list.append(ch['id'])
from tqdm import tqdm
# ========================================
# Load Raw Messages in Channel Info
# ========================================
messages_list = []
for ch_id in tqdm(channel_id_list, desc='loading...'):
    payload = {'token': credentials['channel_api_key'], 'channel': ch_id}
    response = requests.get('https://slack.com/api/channels.history', params=payload)
    if response.status_code == 200:
        json_data = response.json()
        msg_in_ch = {}
        msg_in_ch['channel_id'] = ch_id
        if 'messages' in json_data.keys():
            msg_in_ch['messages'] = json_data['messages']
        else:
            msg_in_ch['messages'] = ''
        messages_list.append(msg_in_ch)
with open('messages_info.json', 'w') as f:
    json.dump(messages_list, f, indent=4, ensure_ascii=False)
```
